```python
import tarfile
import os
from tqdm.auto import tqdm

# Define paths
base_dir = os.path.expanduser("~/Capstone/Implementation/data/mit-supercloud-dataset/gpu/")
tar_path = os.path.join(base_dir, "dual_gpu_0000_parquet_to_0019_parquet_cleaned_split_tensors.tar")

# Extract with progress bar
with tarfile.open(tar_path, 'r') as tar:
    members = tar.getmembers()
    for member in tqdm(iterable=members, total=len(members), desc="Extracting files"):
        tar.extract(member, path=base_dir)

# Delete the archive
os.remove(tar_path)
print("Archive deleted successfully.")
```

```python
import os

def convert_size(size_bytes):
    if size_bytes == 0: return "0 B"
    units = ("B", "KB", "MB", "GB", "TB")
    import math
    i = int(math.floor(math.log(size_bytes, 1024)))
    p = math.pow(1024, i)
    return f"{round(size_bytes / p, 2)} {units[i]}"

def get_folder_stats(target):
    """Calculates sizes and counts for a folder and its contents."""
    total_size = 0
    total_files = 0
    try:
        for root, dirs, files in os.walk(target):
            total_files += len(files)
            for f in files:
                fp = os.path.join(root, f)
                if not os.path.islink(fp):
                    total_size += os.path.getsize(fp)
    except OSError:
        pass
    return total_size, total_files

def print_clean_tree(start_path):
    print(f"{'Structure':<70} | {'Files':>8} | {'Size':>10}")
    print("-" * 95)
    
    # Get the depth of the starting path to calculate relative depth
    start_depth = start_path.rstrip(os.sep).count(os.sep)

    for root, dirs, files in os.walk(start_path):
        # Calculate indentation based on depth
        depth = root.rstrip(os.sep).count(os.sep) - start_depth
        indent = "  " * depth + "|-- " if depth > 0 else ""
        
        folder_name = os.path.basename(root) if depth > 0 else root
        
        # Get stats for the CURRENT folder (including its subfolders)
        size, f_count = get_folder_stats(root)
        
        # Format the line
        folder_label = f"{indent}{folder_name}"
        print(f"{folder_label:<70} | {f_count:>8} | {convert_size(size):>10}")

# --- Run ---
target_dir = "../data/mit-supercloud-dataset/gpu/"
if os.path.exists(target_dir):
    print_clean_tree(target_dir)
else:
    print("Path not found.")
    
```
